```python
import seaborn as sns
import matplotlib.pyplot as plt
def viz(col,file):
  p = sns.countplot(x=col, data=file, palette="muted")
  setup = plt.setp(p.get_xticklabels(), rotation=90)
```

```python
# level of student with respect to raisedhands
student_level = student_record["StageID"]
df = student_record
viz(student_level, df)
```

Middle School was highest in raising hands

```python
# Parent school Satisfaction
ParSatisf = student_record["ParentschoolSatisfaction"]
df = student_record
viz(ParSatisf, df)
```
